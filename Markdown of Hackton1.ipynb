1. 외국인 데이터셋의 전제 조건
  - 한국을 방문하는 외국인들을 조사한 것이다.
  - 다만 관광객으로 간주되지 않는 일부 외국인 입국자(외교관, 군인, 영주권자, 방문 거주자 및 동거자)를 제외한 외국인(한국인 아님)과 재외동포, 승무원 등이 포함되어 있다.
  - 2019년 1월~ 2020년 4월까지의 데이터를 다루고 있고 60개국의 방문객 수를 보여주고 있다.
  - 방문자 수는 성별, 연령, 방문 목적에 따라 구분된다.

2. 60개국을 다 분석할 것인가?
  - 많이 방문한 나라의 수 상위 10개국이나 방문수 변동폭이 큰 나라만을 추출하여 데이터 분석을 하는 것이 좋을 것 같다.
